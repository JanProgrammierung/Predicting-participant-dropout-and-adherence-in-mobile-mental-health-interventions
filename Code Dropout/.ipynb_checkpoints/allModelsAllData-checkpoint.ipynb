```python
import pandas as pd    #import several important packages which give us some useful features
import numpy as np
from sklearn import preprocessing
from numpy import array
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV


#import the library classes for the neural network model
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
from tensorflow.keras import optimizers
from sklearn.datasets import make_circles
from matplotlib import pyplot
from keras.callbacks import EarlyStopping
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

#import class for KNN missing value algorithm
from sklearn.impute import KNNImputer

#import class for MICE missing value algorithm
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

#import class for categorical missing value algorithm
from sklearn.impute import SimpleImputer


#MICE from https://towardsdatascience.com/3-underrated-strategies-to-deal-with-missing-values-a539fb6c0690
#SimpleImputer from https://machinelearningknowledge.ai/how-to-use-sklearn-simple-imputer-simpleimputer-for-filling-missing-values-in-dataset/ and https://dzone.com/articles/imputing-missing-data-using-sklearn-simpleimputer
```

```python
dataFrame = pd.read_excel('DatasetDropout_fromIntroAndBaseline.xlsx') #loading data
dataFrame
```

```python
dataFrame['FilledExit'] = ['0' if x=='no' else '1' for x in dataFrame['FilledExit']]
dataFrame
```

```python
dataFrame.isnull().sum() #compute sum of missing values per column
```

```python
simpleImputer = SimpleImputer (strategy='most_frequent')
dataFrame.gender = simpleImputer.fit_transform(dataFrame['gender'].values.reshape(-1,1))[:,0]
dataFrame.occupation = simpleImputer.fit_transform(dataFrame['occupation'].values.reshape(-1,1))[:,0]
dataFrame.manager = simpleImputer.fit_transform(dataFrame['manager'].values.reshape(-1,1))[:,0]
dataFrame.work = simpleImputer.fit_transform(dataFrame['work'].values.reshape(-1,1))[:,0]
dataFrame.tenure = simpleImputer.fit_transform(dataFrame['tenure'].values.reshape(-1,1))[:,0]
dataFrame.current = simpleImputer.fit_transform(dataFrame['current'].values.reshape(-1,1))[:,0]
dataFrame.frontline = simpleImputer.fit_transform(dataFrame['frontline'].values.reshape(-1,1))[:,0]
dataFrame.change = simpleImputer.fit_transform(dataFrame['change'].values.reshape(-1,1))[:,0]
dataFrame.change_other = simpleImputer.fit_transform(dataFrame['change_other'].values.reshape(-1,1))[:,0]
dataFrame.general_health = simpleImputer.fit_transform(dataFrame['general_health'].values.reshape(-1,1))[:,0]
dataFrame.demand_physical = simpleImputer.fit_transform(dataFrame['demand_physical'].values.reshape(-1,1))[:,0]
dataFrame.demand_private = simpleImputer.fit_transform(dataFrame['demand_private'].values.reshape(-1,1))[:,0]
dataFrame.prev = simpleImputer.fit_transform(dataFrame['prev'].values.reshape(-1,1))[:,0]
```

```python
imputerMICE = IterativeImputer(imputation_order='ascending',max_iter=20,random_state=42,n_nearest_features=None)
imputed_dataFrame = imputerMICE.fit_transform(dataFrame)
```

```python
imputed_dataFrame = pd.DataFrame(imputed_dataFrame)
imputed_dataFrame.columns = ["gender", "age", "occupation", "manager", "work", "tenure", "current", "frontline", "change", "change_other", "general_health", "demand_physical", "demand_private", "prev", "gad", "jcq", "maas", "olbi", "phq", "ptsd", "req", "pssRecoded", "recovery", "SocialSupport", "Control", "smbq", "FilledExit"]
imputed_dataFrame
```

```python
#feature selection

X = imputed_dataFrame.iloc[:,0:26]  #independent columns
y = imputed_dataFrame.iloc[:,-1]    #target column
from sklearn.ensemble import ExtraTreesClassifier
featureModel = ExtraTreesClassifier()
featureModel.fit(X,y)
print(featureModel.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(featureModel.feature_importances_, index=X.columns)
feat_importances.nlargest(20).plot(kind='barh')
pyplot.show()


#statistical feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=26)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Feature','Score']  #naming the dataframe columns
print(featureScores.nlargest(26,'Score'))  #print 10 best features
```

```python
# Define parameters to evaluate

from sklearn import tree

criterion_values         = ['gini', 'entropy']
splitter_values          = ['best', 'random']
max_depth_values         = np.arange(5, 55, 5).tolist()
min_samples_split_values = np.linspace(0.1, 1.0, 10, endpoint=True)
min_samples_leaf_values  = np.linspace(0.1, 0.5, 5, endpoint=True)
max_features_values      = [int(x) for x in np.linspace(1, len(X.columns), num = 18)] #list(range(1,len(X.columns)))
parameters      = {'criterion': criterion_values,
                   'splitter': splitter_values,
                   'max_depth': max_depth_values,
                   'min_samples_split': min_samples_split_values,
                   'min_samples_leaf': min_samples_leaf_values,
                   'max_features': max_features_values,
                   'class_weight': ['balanced']
                  }

dt = tree.DecisionTreeClassifier()

clf = GridSearchCV(      estimator           = dt, 
                         param_grid          = parameters,
                         scoring             = 'roc_auc_ovo_weighted',
                         cv                  = 10,
                         verbose             = 1,
                         n_jobs              = -1
                         )
clf.fit(X,y)
print(clf.best_estimator_)
```

```python
#Here starts the neural network computation

# split into input (X) and output (y) variables

#imputed_dataFrame_norm.to_csv('imputed_dataFrame_norm.csv', header=False)
```

```python
#imputed_array_norm = np.loadtxt('imputed_dataFrame_norm.csv', delimiter=',')

X = np.array(X)
```

```python
model = Sequential()
model.add(Dense(16, input_shape=(X.shape[1],), activation='relu')) # Add an input shape! (features,)
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary() 
```

```python
# compile the model
model.compile(optimizer='Adam', 
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# early stopping callback
# This callback will stop the training when there is no improvement in  
# the validation loss for 10 consecutive epochs.  
#es = EarlyStopping(monitor='val_accuracy', 
 #                                  mode='max', # don't minimize the accuracy!
  #                                 patience=50,
   #                                restore_best_weights=True)

# now we just update our model fit call
history = model.fit(X,
                    y,
    #                callbacks=[es],
                    epochs=200, # you can set this to a big number!
                    batch_size=10,
                    validation_split=0.275,
                    shuffle=True,
                    verbose=1)
```

```python
history_dict = history.history
# Learning curve(Loss)
# let's see the training and validation loss by epoch

# loss
loss_values = history_dict['loss'] # you can change this
val_loss_values = history_dict['val_loss'] # you can also change this

# range of X (no. of epochs)
epochs = range(1, len(loss_values) + 1) 

# plot
plt.plot(epochs, loss_values, 'bo', label='Training loss')
plt.plot(epochs, val_loss_values, 'orange', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Learning curve(accuracy)
# let's see the training and validation accuracy by epoch

# accuracy
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

# range of X (no. of epochs)
epochs = range(1, len(acc) + 1)

# plot
# "bo" is for "blue dot"
plt.plot(epochs, acc, 'bo', label='Training accuracy')
# orange is for "orange"
plt.plot(epochs, val_acc, 'orange', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# this is the max value - should correspond to
# the HIGHEST train accuracy
np.max(val_acc)
```

```python
# see how these are numbers between 0 and 1? 
model.predict(X) # prob of successes (survival)
np.round(model.predict(X),0) # 1 and 0 (survival or not)

# so we need to round to a whole number (0 or 1),
# or the confusion matrix won't work!
preds = np.round(model.predict(X),0)

# confusion matrix
mat = confusion_matrix(y, preds)
sns.heatmap(mat.T, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values
plt.show() 


print(classification_report(y, preds))
```
