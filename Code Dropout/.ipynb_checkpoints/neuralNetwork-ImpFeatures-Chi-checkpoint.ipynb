```python
import pandas as pd    #import several important packages which give us some useful features
import numpy as np
from sklearn import preprocessing
from numpy import array
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf

#import the library classes for the neural network model
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras import optimizers
from sklearn.datasets import make_circles
from matplotlib import pyplot

#import class for KNN missing value algorithm
from sklearn.impute import KNNImputer

#import class for MICE missing value algorithm
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

#import class for categorical missing value algorithm
from sklearn.impute import SimpleImputer


#MICE from https://towardsdatascience.com/3-underrated-strategies-to-deal-with-missing-values-a539fb6c0690
#SimpleImputer from https://machinelearningknowledge.ai/how-to-use-sklearn-simple-imputer-simpleimputer-for-filling-missing-values-in-dataset/ and https://dzone.com/articles/imputing-missing-data-using-sklearn-simpleimputer
```

```python
dataFrame = pd.read_excel('DatasetDropout_fromIntroAndBaseline.xlsx') #loading data
dataFrame
```

```python
dataFrame['FilledExit'] = ['0' if x=='no' else '1' for x in dataFrame['FilledExit']]
dataFrame
```

```python
dataFrameChi = dataFrame[['gender', 'age', 'current', 'frontline', 'occupation', 'tenure', 'work', 'change', 'prev', 'gad', 'maas', 'phq', 'req', 'FilledExit']]
```

```python
dataFrameChi.isnull().sum() #compute sum of missing values per column
```

```python
simpleImputer = SimpleImputer (strategy='most_frequent')
dataFrameChi.gender = simpleImputer.fit_transform(dataFrameChi['gender'].values.reshape(-1,1))[:,0]
dataFrameChi.occupation = simpleImputer.fit_transform(dataFrameChi['occupation'].values.reshape(-1,1))[:,0]
dataFrameChi.current = simpleImputer.fit_transform(dataFrameChi['current'].values.reshape(-1,1))[:,0]
dataFrameChi.frontline = simpleImputer.fit_transform(dataFrameChi['frontline'].values.reshape(-1,1))[:,0]
dataFrameChi.tenure = simpleImputer.fit_transform(dataFrameChi['tenure'].values.reshape(-1,1))[:,0]
dataFrameChi.work = simpleImputer.fit_transform(dataFrameChi['work'].values.reshape(-1,1))[:,0]
dataFrameChi.change = simpleImputer.fit_transform(dataFrameChi['change'].values.reshape(-1,1))[:,0]
dataFrameChi.prev = simpleImputer.fit_transform(dataFrameChi['prev'].values.reshape(-1,1))[:,0]



```

```python
dataFrameChi
```

```python
dataFrameChi.isnull().sum() #compute sum of missing values per column
```

```python
imputerMICE = IterativeImputer(imputation_order='ascending',max_iter=10,random_state=42,n_nearest_features=None)
imputed_dataFrameChi = imputerMICE.fit_transform(dataFrameChi)
```

```python
imputed_dataFrameChi = pd.DataFrame(imputed_dataFrameChi)
imputed_dataFrameChi.columns = ['gender', 'age', 'current', 'frontline', 'occupation', 'tenure', 'work', 'change', 'prev', 'gad', 'maas', 'phq', 'req', 'FilledExit']
imputed_dataFrameChi
```

```python
#feature selection

X = imputed_dataFrameChi.iloc[:,0:13]  #independent columns
y = imputed_dataFrameChi.iloc[:,-1]    #target column
from sklearn.ensemble import ExtraTreesClassifier
featureModel = ExtraTreesClassifier()
featureModel.fit(X,y)
print(featureModel.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(featureModel.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
pyplot.show()


#statistical feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=13)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Feature','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

```python
# generate 2d classification dataset
X, y = make_circles(n_samples=1000, noise=0.1, random_state=1)
# split into train and test
n_train = 500
trainX, testX = X[:n_train, :], X[n_train:, :]
trainy, testy = y[:n_train], y[n_train:]
```

```python
# define the keras model

model = Sequential()

model.add(Dense(50, input_dim=2, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(1, activation='sigmoid'))
opt = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
```

```python
# compile the keras model

model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
# fit model
history = model.fit(trainX, trainy, validation_data=(testX, testy), epochs=200, verbose=0)
# evaluate the model
_, train_acc = model.evaluate(trainX, trainy, verbose=0)
_, test_acc = model.evaluate(testX, testy, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
# plot loss during training
pyplot.subplot(211)
pyplot.title('Loss')
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
# plot accuracy during training
pyplot.subplot(212)
pyplot.title('Accuracy')
pyplot.plot(history.history['accuracy'], label='train')
pyplot.plot(history.history['val_accuracy'], label='test')
pyplot.legend()
pyplot.show()
```

```python
# evaluate the keras mode

_, mse = model.evaluate(X, y)
print('MSE: %.2f' % (mse*100))
```
