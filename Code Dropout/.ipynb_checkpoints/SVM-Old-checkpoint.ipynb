```python
import pandas as pd    #import several important packages which give us some useful features
import numpy as np
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pickle #library for saving our model
```

```python
svmModel = pd.read_excel('GBIntroNoEmpty.xlsx') #loading data
svmModel
```

```python
svmModelEdited = svmModel[['gender', 'age', 'manager', 'general_health', 'demand_physical', 'demand_private', 'gad', 'maas', 'olbi', 'phq', 'ptsd', 'pssRecoded', 'Dropout']]
svmModelEdited
```

```python
svmModelEdited.isnull().sum() #compute sum of missing values per column
```

```python
#Normalisation
x = svmModelEdited.values
```

```python
print (x)
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
svmModelEdited = svmModelEdited.copy()
svmModelEditedNorm = pd.DataFrame(x_scaled, columns=svmModelEdited.columns)
```

```python
svmModelEditedNorm.head(len(svmModelEditedNorm))
```

```python
X = svmModelEditedNorm.drop('Dropout', axis=1)
y = svmModelEditedNorm['Dropout']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = tree.DecisionTreeClassifier(random_state=42, criterion='entropy')
clf = clf.fit(X_train, y_train)
```

```python
tree.plot_tree(clf.fit(X_train, y_train)) 
```

```python
ypred = clf.predict(X_test)
ypred
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, ypred))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ypred, y_test)
sns.heatmap(mat.T, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values
plt.show() 
```

```python
forest = RandomForestClassifier(n_estimators=1000, max_depth=5, criterion='entropy',
                        random_state=42)
forest.fit(X_train, y_train)
```

```python
ypredbag = forest.predict(X_test)
```

```python
print(metrics.classification_report(y_test, ypredbag))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, ypredbag)
sns.heatmap(mat.T, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t) # update the ylim(bottom, top) values
plt.show() 
```
